Problem 5: The password strength check

```python
pwd = 'hello@2026'
l = len(pwd)
is_length = l>=8
print('Password:', pwd)
print('Length OK (>=8):', is_length)
has_char = '@' in pwd
print("Contains '@':", has_char)
is_strong = is_length and has_char
print('Strong:', is_strong)
```
